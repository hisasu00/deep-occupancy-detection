# Import Module

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Free Params

```python
num_obs = 48
num_days = 5
num_samples = num_obs*num_days

energy_1 = 1e6
energy_0 = 0.2*1e6

breakfast_t = 5
dinner_t = 36
```

# Generate Data

```python
test_x = []
test_y = []

for t in range(0, num_obs, 1):
    if t < breakfast_t:
        test_x.append(energy_1)
        test_y.append(1)
    elif dinner_t < t:
        test_x.append(energy_1)
        test_y.append(1)
    else:
        test_x.append(energy_0)
        test_y.append(0)

test_x = test_x*num_days
test_y = test_y*num_days
noise = np.random.normal(energy_1/25, energy_1/10, (num_samples, ))
```

# Energy

```python
plt.plot(test_x + noise)
plt.xticks(range(1, num_samples, 47))
plt.ylim([0, 2.5e6])
plt.show()
```
